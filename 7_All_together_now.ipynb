## Notebook 7: Analysing lots of seismograms

You're now in a position to analyse seismograms.  Try each of the following.

### 1. Analyse an earthquake in multiply frequency bands

A. Pick an M>8 earthquake. (earthquake.usgs.gov has a nicely accessible catalogue.)

B. Identify a station with seismic data that's a few hundred km away.

c. Download around half an hour of data around the earthquake.

D. Remove the instrument response.

E. Detrend, taper, and bandpass filter between 0.1 and 1 Hz.

F. Compute the amplitude through time.

G. Lowpass filter the amplitude time series below 0.1 Hz.

H. Resample the amplitude time series so that there is only one point per second.

I. Redo steps E-G for a few more frequency bands.

J. Compare the observations.  Can you understand the variations in each frequency band?

K. Try a few more, smaller earthquakes in a similar location.  How do their frequency contents differ from those of the M>8 earthquake?

### 2. A week of seismic amplitude variation

A. Choose a station where you'd like to examine seismic amplitude variations over a week or more.

B. Pick a week and divide it into half-hour intervals.

C. For each interval
- download the data
- remove the instrument response
- bandpass filter between 1 and 10 Hz
- compute the amplitude through time
- lowpass filter the amplitude time series below 0.001 Hz
- resample the amplitude time series so there is only one point per 100 s
- trim off the ends
- save the amplitude in a sensible location

D. Load and merge the amplitude time series to get a week-long record.

E. Lowpass filter the merged record to isolate periods longer than 3 hours (to < 1/10800 Hz).

F. Examine the amplitude through time.  What do you think causes the patterns you observe?

F. Try a similar analysis in another frequency band (not 1 to 10 Hz).  You can make this more efficient by downloading the data once, copying it, and filtering the various copies differently.
